```
import bs4
import requests
```

```
url="https://www.daraz.pk/headphones-headsets/?spm=a2a0e.pdp.cate_2_2.1.5a5e2191WllAad"
```

```
data=requests.get(url)
```

```
soup=bs4.BeautifulSoup(data.text,"html.parser")
```

```
print(soup.prettify())
```

```
for para in soup.find("p"):
    print(para)
```

```
for para in soup.find_all("p"):
    print(para)
```

```
for para in soup.find_all("p"):
    print(para.text)
```

```
for links in soup.find_all("a"):
    link =links.get("href")
    print (link)
```

```
for links in soup.find_all("a"):
    link =str(links.get("href"))
    if link [0:2] == "//"  :
        c=("http:"+ link)
        print (c)    
    elif link [0:2] == "None" :
        print("http:"+ link)
```

```
for ps in soup.find_all("li"):
    print(ps.text)
```
